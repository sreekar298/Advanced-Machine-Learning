# Question 4: Linear Regression (10 pts)

1. What is the difference between R-square and adjusted R square and why is it desirable to use the adjusted value? (4 pts)

2. Overfitting usually happens in complex models. Linear Regression is a fairly simple model. Could overfitting happen in Linear Regression? If so, please explain the scenario in which it could happen and how we can tackle it. (6 pts)


#### Solution
1. R square indicates the amount of variability that is explained by the variables in the model. We would expect that to be as high as possible. It is calculated by : 
$$ {1} - {RSS \over TSS} $$ 
where RSS $ = $ sum of squared residuals i.e sum of (yactual - ypred)^2 
TSS is the total sum of squared residuals i.e sum of (yactual - ymean)^2. Ypred is given by B0 + B1x. 

As we increase the number of variables in our model, the RSS goes down, i.e the R square value keeps increasing. But in multiple regression when we have a lot of variables, variable selection becomes important and increasing the number of variables in the model doesn't lead to a significant increase in R square. Which indicates that we're just adding noise to the model. To avoid this noise, we use the adjusted R square value which is a better estimate of how much variablity is explained by the selected set of variables in the model. The intuition behind the adjusted R2 is that once all of the correct variables have been included in the model, adding additional noise variables will lead to only a very small decrease in RSS. 

2. Yes, overfitting can happen in a linear regression when it's made into a complex model than it's supposed to be. Having too many terms than the number of obeservations would lead to overfitting and the terms in the equation would start representing noise and not the actual relationship in the population. In order to overcome this, one should pick a sample size that's large enough for the number of variables. Another way would be to reduce the variables that explain the noise and select only a subset of the variables.

#### Part 4 B) Pros and cons of Ridge, Lasso and ElasticNet

**Ridge Regression**

Pros
- Ridge regression helps prevent overfitting
- Trades variance for bias (i.e. in presence of co-linearity, it is worth to have biased results, in order to lower the variance.)

Cons
- Increases bias
- It doesn't squish any of coefficients to 0, therefore variable selection is not possible 

**Lasso Regression**

Pros
- Select features, by shrinking co-efficient towards zero.
- Lasso also helps prevent overfitting

Cons
- LASSO will select only one feature from a group of correlated features, the selection is arbitrary in nature.
- It's prediction performance is worse than Ridge

**ElasticNet**

Pros
- It's a mix of the two models, i.e combines feature elimination from Lasso and feature coefficient reduction from the Ridge model to improve your models predictions.
- Computationally more expensive

#### Part 7) Why did we have to scale the data in ridge and lasso regression?

The reason to scale the data before regularization is to make sure that the regularization term  has a similar effect on the variable involved. We must scale the data before regularization since regularization penalizes the coefficient weights. When we directly apply regularization where we have two features with their values ranging with significant difference i.e one being extremely small and the other being extremely large then it is going to be unfair with the small range. Hence we do scaling to avoid this unfair penalization. 

Here, we see that when we defined the model using unscaled data, the best alpha that we saw was very low. This might be because since the actual feature values are very large, in order to minimise the loss function with the regularization term, we need a very small alpha.
